```python
def simulate_ball(probs, runs, wickets):
    
    RV = np.random.uniform()*100
    for i in range(1, len(probs)):
        if(RV < probs[i] and RV >= probs[i-1]):
            res = i-1

    if(res == 0):
        result = 'dot'
    elif(res == 1):
        result =  'one'
        runs += 1
    elif(res == 2):
        result =  'two'
        runs += 2
    elif(res == 3):
        result =  'three'
        runs += 3
    elif(res == 4):
        result =  'four'
        runs += 4
    elif(res == 5):
        result =  'five'
        runs += 5
    elif(res == 6):
        result =  'six'
        runs += 6
    elif(res == 7):
        result =  'extra run'
        runs += calculate_extra_runs()
    elif(res == 8):
        result =  'wicket'
        runs += calculate_wicket_runs()
        wickets +=1
    return result, runs, wickets
```

```python
def calculate_extra_runs():
    extra_probs = [84.8034606 ,  6.75192778,  1.20368629,  3.40417529,  3.3101373 ,
        0.        ,  0.52661275]
    for i in range(1, len(extra_probs)):
        extra_probs[i] = extra_probs[i] + extra_probs[i-1]

    extra_probs.insert(0, 0)

    RV = np.random.uniform()*100
    for i in range(1, len(extra_probs)):
        if(RV < extra_probs[i] and RV >= extra_probs[i-1]):
            return i
    return 0
```

```python
def calculate_wicket_runs():
    wicket_probs = [89.1104006 ,  3.59225127,  0.16926838]
    for i in range(1, len(wicket_probs)):
        wicket_probs[i] = wicket_probs[i] + wicket_probs[i-1]

    wicket_probs.insert(0, 0)

    RV = np.random.uniform()*100
    for i in range(1, len(wicket_probs)):
        if(RV < wicket_probs[i] and RV >= wicket_probs[i-1]):
            return i
    return 0
```

```python
def simulate_innings(probs):
    wickets = 0
    runs = 0
    for o in range(1,21):
        for b in range(1,7):
            if(wickets==10):
                break
            res,runs, wickets = simulate_ball(probs, runs, wickets)
            if(res == "extra run"):
                b -= 1
            #print("" + str(o) + ":" + str(b) + ":" + str(res))
        if(wickets == 10):
            break
        
    return runs, wickets
```

```python
def simulate_game(probs1, probs2):
    runs1 = 0
    wickets1 = 0
    runs2 = 0
    wickets2 = 0

    runs1, wickets1 = simulate_innings(probs1)
    runs2, wickets2 = simulate_innings(probs2)

    # print("----- End of Match -----")
    # print("Team 1 score: " + str(runs1) + "/" + str(wickets1))
    # print("Team 2 score: " + str(runs2) + "/" + str(wickets2))

    if(runs1 > runs2):
        return "Team 1 wins", runs1, wickets1, runs2, wickets2 
    elif(runs1 < runs2):
        return "Team 2 wins", runs1, wickets1, runs2, wickets2 
    else:
        return "Draw", runs1, wickets1, runs2, wickets2 
```

```python
def test_func(probs1, probs2, n):
    runs = []
    wickets = []
    team1_wins = 0
    team2_wins = 0
    
    for i in range(n):
        res, runs1, wickets1, runs2, wickets2 = simulate_game(probs1, probs2)
        runs.append(runs1)
        runs.append(runs2)
        wickets.append(wickets1)
        wickets.append(wickets2)
        if(res == "Team 1 wins"):
            team1_wins += 1
        elif(res == "Team 2 wins"):
            team2_wins += 1
        #print(res)
    
    return sum(runs)/len(runs), sum(wickets)/len(wickets), team1_wins*100/n, team2_wins*100/n
```

```python
def calculate_inning1_prob():
    T20_innings1_data = IPL_ball_data.loc[IPL_ball_data['inning'] == 1]
    T20_innings1_data = T20_innings1_data.reset_index(drop=True)
    
    T20_runs = T20_innings1_data.loc[(T20_innings1_data['extra_runs'] == 0) & (T20_innings1_data['is_wicket'] == 0)]
    T20_extra = T20_innings1_data.loc[(T20_innings1_data['extra_runs'] > 0) & (T20_innings1_data['is_wicket'] == 0)]
    T20_wicket = T20_innings1_data.loc[T20_innings1_data['is_wicket'] != 0]
    
    innings1_results_table = pd.DataFrame({'Outcome': pd.Series(dtype='str'),
                   'T20_prob': pd.Series(dtype='float')})
    for i in sorted(T20_runs['total_runs'].unique()):
    #print("Probability of scoring " + str(i) + " run : " + str(T20_data['batsman_runs'].value_counts()[i]/T20_data.shape[0]*100))
        innings1_results_table = innings1_results_table.append({'Outcome': str(i), 'T20_prob': round(T20_runs['batsman_runs'].value_counts()[i]/T20_innings1_data.shape[0]*100,7)},ignore_index=True)

    innings1_results_table = innings1_results_table.append({'Outcome': 'extra runs', 'T20_prob': T20_extra.shape[0]/T20_innings1_data.shape[0]*100},ignore_index=True)
    innings1_results_table = innings1_results_table.append({'Outcome': 'wicket', 'T20_prob':T20_wicket.shape[0]/T20_innings1_data.shape[0]* 100},ignore_index=True)
    return innings1_results_table
    
```

```python
def calculate_matchup_prob(batting_team, bowling_team):
    batting_data = IPL_ball_data.loc[(IPL_ball_data['batting_team']==batting_team) & (IPL_ball_data['bowling_team']==bowling_team)]
    
    runs_data = batting_data.loc[(batting_data['extra_runs'] == 0) & (batting_data['is_wicket'] == 0)]
    extra_data = batting_data.loc[(batting_data['extra_runs'] > 0) & (batting_data['is_wicket'] == 0)]
    wicket_data = batting_data.loc[batting_data['is_wicket'] != 0]
    #print(noextra_data)

    batting_table = pd.DataFrame({'Outcome': pd.Series(dtype='str'),
                    'T20_prob': pd.Series(dtype='float')})
    for i in range(7):
        if i in runs_data['total_runs'].unique():
            batting_table = batting_table.append({'Outcome': str(i), 'T20_prob': runs_data['total_runs'].value_counts()[i]/batting_data.shape[0]*100},ignore_index=True)
        else:
            batting_table = batting_table.append({'Outcome': str(i), 'T20_prob': 0},ignore_index=True)
       
    batting_table = batting_table.append({'Outcome': 'extra runs', 'T20_prob': extra_data.shape[0]/batting_data.shape[0]*100},ignore_index=True)
    batting_table = batting_table.append({'Outcome': 'wicket', 'T20_prob': wicket_data.shape[0]/batting_data.shape[0]* 100},ignore_index=True)
    #print(runs_data)
    return batting_table
    
```

```python
def play_matchup(team1, team2, n):
    
    
    probs1 = calculate_matchup_prob(team1, team2)['T20_prob'].values.tolist()
    probs2 = calculate_matchup_prob(team2, team1)['T20_prob'].values.tolist()
    for i in range(1, len(probs1)):
        probs1[i] = probs1[i] + probs1[i-1]
    for i in range(1, len(probs2)):
        probs2[i] = probs2[i] + probs2[i-1]

    probs1.insert(0, 0)
    probs2.insert(0, 0)
    
    #print(calculate_matchup_prob(team1, team2))
    #print(calculate_matchup_prob(team2, team1))
    
    runs1_arr = []
    wickets1_arr = []
    runs2_arr = []
    wickets2_arr = []
    team1_wins = 0
    team2_wins = 0
    
    for i in range(n):
        res, runs1, wickets1, runs2, wickets2 = simulate_game(probs1, probs2)
        runs1_arr.append(runs1)
        runs2_arr.append(runs2)
        wickets1_arr.append(wickets1)
        wickets2_arr.append(wickets2)
        if(res == "Team 1 wins"):
            team1_wins += 1
        elif(res == "Team 2 wins"):
            team2_wins += 1
        #print(res)
        
    avg_runs1 = sum(runs1_arr)/len(runs1_arr)
    avg_wickets1 = sum(wickets1_arr)/len(wickets1_arr)
    avg_runs2 = sum(runs2_arr)/len(runs2_arr)
    avg_wickets2 = sum(wickets2_arr)/len(wickets2_arr)
    team1_w_p = team1_wins*100/n
    team2_w_p = team2_wins*100/n
    
    print("Average runs " + team1 + ": " +  str(avg_runs1))
    print("Average wickets " + team1 + ": " + str(avg_wickets1))
    print("Average runs " + team1 + ": " +  str(avg_runs2))
    print("Average wickets " + team2 + ": " + str(avg_wickets2))
#     print("% won by " + team1 + ": " + str(team1_w_p) + "%")
#     print("% won by " + team2 + ": " + str(team2_w_p) + "%")

    verify_matchup_results(team1, team2)
```
