```python
import pandas as pd  
import numpy as np
```

```python
df_2008 = pd.read_csv("IPL_Joint2008.csv")
df_2009 = pd.read_csv("IPL_Joint2009.csv")
df_2010 = pd.read_csv("IPL_Joint2010.csv")
df_2011 = pd.read_csv("IPL_Joint2011.csv")
df_2012 = pd.read_csv("IPL_Joint2012.csv")
df_2013 = pd.read_csv("IPL_Joint2013.csv")
df_2014 = pd.read_csv("IPL_Joint2014.csv")
df_2015 = pd.read_csv("IPL_Joint2015.csv")
df_2016 = pd.read_csv("IPL_Joint2016.csv")
df_2017 = pd.read_csv("IPL_Joint2017.csv")
df_2018 = pd.read_csv("IPL_Joint2018.csv")
df_2019 = pd.read_csv("IPL_Joint2019.csv")
df_2020 = pd.read_csv("IPL_Joint2020.csv")
df_2021 = pd.read_csv("IPL_Joint2021.csv")
df_2022 = pd.read_csv("IPL_Joint2022.csv")
```

```python
df_2008.insert(0,"Season",2008)
df_2009.insert(0,"Season",2009)
df_2010.insert(0,"Season",2010)
df_2011.insert(0,"Season",2011)
df_2012.insert(0,"Season",2012)
df_2013.insert(0,"Season",2013)
df_2014.insert(0,"Season",2014)
df_2015.insert(0,"Season",2015)
df_2016.insert(0,"Season",2016)
df_2017.insert(0,"Season",2017)
df_2018.insert(0,"Season",2018)
df_2019.insert(0,"Season",2019)
df_2020.insert(0,"Season",2020)
df_2021.insert(0,"Season",2021)
df_2022.insert(0,"Season",2022)
```

```python
df = pd.concat([df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022])
df
```

### Remove unnamed Column

```python
df.columns.values
```

```python
df = df.reset_index(drop=True)
df = df.drop(['Unnamed: 0'], axis = 1)
```

### Remove duplicate teams (Handling changes in team name)

```python
df['winner'].unique()
```

```python
df['winner'].replace('Deccan Chargers', 'Sunrisers Hyderabad', inplace=True)
df['winner'].replace('Rising Pune Supergiants', 'Pune Warriors', inplace=True)
df['winner'].replace('Gujarat Lions', 'Gujarat Titans', inplace=True)
df['winner'].replace('Kings XI Punjab', 'Punjab Kings', inplace=True)

df['HomeTeam'].replace('Deccan Chargers', 'Sunrisers Hyderabad', inplace=True)
df['HomeTeam'].replace('Rising Pune Supergiants', 'Pune Warriors', inplace=True)
df['HomeTeam'].replace('Gujarat Lions', 'Gujarat Titans', inplace=True)
df['HomeTeam'].replace('Kings XI Punjab', 'Punjab Kings', inplace=True)

df['AwayTeam'].replace('Deccan Chargers', 'Sunrisers Hyderabad', inplace=True)
df['AwayTeam'].replace('Rising Pune Supergiants', 'Pune Warriors', inplace=True)
df['AwayTeam'].replace('Gujarat Lions', 'Gujarat Titans', inplace=True)
df['AwayTeam'].replace('Kings XI Punjab', 'Punjab Kings', inplace=True)

df['toss_winner'].replace('Deccan Chargers', 'Sunrisers Hyderabad', inplace=True)
df['toss_winner'].replace('Rising Pune Supergiants', 'Pune Warriors', inplace=True)
df['toss_winner'].replace('Gujarat Lions', 'Gujarat Titans', inplace=True)
df['toss_winner'].replace('Kings XI Punjab', 'Punjab Kings', inplace=True)
```
