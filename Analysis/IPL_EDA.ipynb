```python
IPL_data = IPL_data.loc[(IPL_data['Season']!=2021) & (IPL_data['Season']!=2022)]
IPL_data
```

### Most wins in each season

```python
teams_per_season = IPL_data.groupby('Season')['Winner'].value_counts()
year = 2008
win_per_season_df = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in teams_per_season.iteritems():    
    if items[0][0]==year:
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        win_per_season_df = win_per_season_df.append(win_series)
        year += 1
win_per_season_df = win_per_season_df.reset_index(drop=True)
win_per_season_df
```

### Matches per venue

```python
venue_ser = IPL_data['Venue'].value_counts()
venue_df = pd.DataFrame(columns=['venue', 'matches'])
for items in venue_ser.iteritems():
    temp_df = pd.DataFrame({
        'venue':[items[0]],
        'matches':[items[1]]
    })
    venue_df = venue_df.append(temp_df, ignore_index=True)
    
plt.title("IPL Venues")
sns.barplot(x='matches', y='venue', data=venue_df)
venue_df.head(10)
```

```python
top_venues = venue_df['venue'].head(10).values
top_venues
```

### Average Scores in each stadium

```python
venue_df = pd.DataFrame()
venue_df['Stadium'] = IPL_data['Venue']
venue_df['Score'] = (IPL_data['HomeRuns'] + IPL_data['AwayRuns'])/2
venue_df['Wickets'] = (IPL_data['HomeWickets'] + IPL_data['AwayWickets'])/2
venue_df['Overs'] = (IPL_data['HomeOvers'] + IPL_data['AwayOvers'])/2

grouped_venue_df = venue_df.groupby('Stadium').mean()
grouped_venue_df.loc[top_venues]
```

### Teams with most wins

```python
team_wins_ser = IPL_data['Winner'].value_counts()

team_wins_df = pd.DataFrame(columns=["team", "wins"])
for items in team_wins_ser.iteritems():
    temp_df1 = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    team_wins_df = team_wins_df.append(temp_df1, ignore_index=True)

plt.title("Total Victories of IPL Teams")

sns.barplot(x='wins', y='team', data=team_wins_df, palette='Paired')
#ax.bar_label(ax.containers[0])

team_wins_df
```

### MVP

```python
mvp_ser = IPL_data['PlayerOfMatch'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    if count>9:
        break
    else:
        temp_df2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
        mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
        count += 1 
plt.title("Top Ten IPL Players")
sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='Paired');
mvp_ten_df
```

### Luckiest Team

```python
toss_ser = IPL_data['TossWinner'].value_counts()

toss_df = pd.DataFrame(columns=["team", "wins"])

for items in toss_ser.iteritems():
    temp_df3 = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    toss_df = toss_df.append(temp_df3, ignore_index=True) 
plt.title("How IPL Teams fared in toss?")
sns.barplot(x='wins', y='team', data=toss_df, palette='Paired');
toss_df
```

### Home vs Away

```python
home_away_df = pd.DataFrame(data = {'Stats': ['Runs','Overs','Wickets'], 'Home': [IPL_data["HomeRuns"].mean(), IPL_data["HomeOvers"].mean(), IPL_data["HomeWickets"].mean()], 'Away': [IPL_data["AwayRuns"].mean(), IPL_data["AwayOvers"].mean(), IPL_data["AwayWickets"].mean()], 'Total': [(IPL_data["HomeRuns"].mean() + IPL_data["AwayRuns"].mean())/2, (IPL_data["HomeOvers"].mean() + IPL_data["AwayOvers"].mean())/2, (IPL_data["HomeWickets"].mean() + IPL_data["AwayWickets"].mean())/2]})
home_away_df
```

### Team Comparison Home vs Away

```python
home_df = pd.DataFrame()
home_df['Team'] = IPL_data['HomeTeam']
home_df['HomeWin%'] = np.where(IPL_data['HomeTeam'] == IPL_data['Winner'], 1, 0)
grouped_df = home_df.groupby('Team').mean()

away_df = pd.DataFrame()
away_df['Team'] = IPL_data['AwayTeam']
away_df['AwayWin%'] = np.where(IPL_data['AwayTeam'] == IPL_data['Winner'], 1, 0)
grouped_away_df = away_df.groupby('Team').mean()

grouped_df['AwayWin%'] = grouped_away_df['AwayWin%']


del home_df
del away_df
del grouped_away_df

fig = plt.figure(figsize = (15, 5))
x = np.arange(grouped_df.shape[0])
width = 0.3
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, grouped_df['HomeWin%'], width, label='HomeWin%')
plt.bar(x+0.2, grouped_df['AwayWin%'], width, label='AwayWin%')
plt.xticks(x, ['CSK', 'DC', 'GT', 'KTK', 'KKR', 'LSG', 'MI', 'PW', 'PK', 'RR', 'RCB', 'SH'])
plt.xlabel("IPL Teams")
plt.ylabel("Win %")
plt.legend()
plt.show()
```

### First Innings vs Second Innings

```python
temp_df = pd.DataFrame()
temp_df['Innings1_score'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['HomeRuns'], IPL_data['AwayRuns']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['HomeRuns'], IPL_data['AwayRuns']))
temp_df['Innings2_score'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['AwayRuns'], IPL_data['HomeRuns']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['AwayRuns'], IPL_data['HomeRuns']))                                                            

temp_df['Innings1_wickets'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['HomeWickets'], IPL_data['AwayWickets']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['HomeWickets'], IPL_data['AwayWickets']))
temp_df['Innings2_wickets'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['AwayWickets'], IPL_data['HomeWickets']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['AwayWickets'], IPL_data['HomeWickets']))                                                            

temp_df['Innings1_overs'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['HomeOvers'], IPL_data['AwayOvers']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['HomeOvers'], IPL_data['AwayOvers']))
temp_df['Innings2_overs'] = np.where(IPL_data['TossDecision'] == 'bat', 
                                       np.where(IPL_data['HomeTeam'] == IPL_data['TossWinner'], IPL_data['AwayOvers'], IPL_data['HomeOvers']), 
                                       np.where(IPL_data['AwayTeam'] == IPL_data['TossWinner'], IPL_data['AwayOvers'], IPL_data['HomeOvers']))                                                            


temp_df.head()
```

```python
print("Average score on innings 1 : " + str(temp_df['Innings1_score'].mean()))
print("Average score on innings 2 : " + str(temp_df['Innings2_score'].mean()))
print()

print("Average wickets on innings 1 : " + str(temp_df['Innings1_wickets'].mean()))
print("Average wickets on innings 2 : " + str(temp_df['Innings2_wickets'].mean()))
print()

print("Average overs on innings 1 : " + str(temp_df['Innings1_overs'].mean()))
print("Average overs on innings 2 : " + str(temp_df['Innings2_overs'].mean()))
print()
```
