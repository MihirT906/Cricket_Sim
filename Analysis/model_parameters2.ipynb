```python
import pandas as pd  
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
T20_data = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
T20_data = T20_data.sort_values(['id', 'inning', 'over', 'ball'])
T20_data = T20_data.reset_index(drop=True)
T20_data.head()
```

```python
T20_data.shape
```

```python
T20_innings1_data = T20_data.loc[T20_data['inning'] == 1]
T20_innings1_data = T20_innings1_data.reset_index(drop=True)
T20_innings1_data
```

### Innings 1 table

```python
T20_runs = T20_innings1_data.loc[(T20_innings1_data['extra_runs'] == 0) & (T20_innings1_data['is_wicket'] == 0)]
T20_extra = T20_innings1_data.loc[(T20_innings1_data['extra_runs'] > 0) & (T20_innings1_data['is_wicket'] == 0)]
T20_wicket = T20_innings1_data.loc[T20_innings1_data['is_wicket'] != 0]
```

```python
innings1_results_table = pd.DataFrame({'Outcome': pd.Series(dtype='str'),
                   'T20_prob': pd.Series(dtype='float')})
for i in range(7):
    #print("Probability of scoring " + str(i) + " run : " + str(T20_data['batsman_runs'].value_counts()[i]/T20_data.shape[0]*100))
    innings1_results_table = innings1_results_table.append({'Outcome': str(i), 'T20_prob': T20_runs['batsman_runs'].value_counts()[i]/T20_innings1_data.shape[0]*100},ignore_index=True)

innings1_results_table = innings1_results_table.append({'Outcome': 'extra runs', 'T20_prob': T20_extra.shape[0]/T20_innings1_data.shape[0]*100},ignore_index=True)
innings1_results_table = innings1_results_table.append({'Outcome': 'wicket', 'T20_prob':T20_wicket.shape[0]/T20_innings1_data.shape[0]* 100},ignore_index=True)
innings1_results_table
```

```python
innings1_results_table['T20_prob'].values
```

```python
extras_table = pd.DataFrame({'Runs Scored': pd.Series(dtype='str'),
                   'prob': pd.Series(dtype='float')})
for i in range(1,8):
    if(i==6):
        extras_table = extras_table.append({'Runs Scored': str(i), 'prob': 0},ignore_index=True)
    else:
    #print("Probability of scoring " + str(i) + " run : " + str(T20_data['batsman_runs'].value_counts()[i]/T20_data.shape[0]*100))
        extras_table = extras_table.append({'Runs Scored': str(i), 'prob': T20_extra['total_runs'].value_counts()[i]/T20_extra.shape[0]*100},ignore_index=True)

extras_table
```

```python
extras_table['prob'].values
```

```python
wickets_table = pd.DataFrame({'Runs Scored': pd.Series(dtype='str'),
                   'prob': pd.Series(dtype='float')})
for i in range(3):
#     if(i==6):
#         wickets_table = wickets_table.append({'Runs Scored': str(i), 'prob': 0},ignore_index=True)
#     else:
    #print("Probability of scoring " + str(i) + " run : " + str(T20_data['batsman_runs'].value_counts()[i]/T20_data.shape[0]*100))
        wickets_table = wickets_table.append({'Runs Scored': str(i), 'prob': T20_wicket['total_runs'].value_counts()[i]/T20_extra.shape[0]*100},ignore_index=True)

wickets_table
```
