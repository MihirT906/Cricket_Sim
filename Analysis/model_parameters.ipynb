### Using ball-by-ball data to calculate avg score

```python
#ODI
score_sum = 0;
match_count = 0;
for i in range(1,ODI_data.shape[0]):
    if(ODI_data['Score'][i-1]>ODI_data['Score'][i]):
        score_sum = score_sum+ODI_data['Score'][i-1]
        match_count=match_count+1
print("Average score in ODI using ball by ball data: " + str(score_sum/match_count))
```

```python
#IPL
inning_score = 0;
score_sum = 0;
match_count = 0;
for i in range(1,T20_data.shape[0]):
    if(T20_data['over'][i-1] <= T20_data['over'][i]):
        inning_score = inning_score + T20_data['total_runs'][i-1]
    else:
        inning_score = inning_score + T20_data['total_runs'][i-1]
        score_sum = score_sum+inning_score
        #print(inning_score)
        inning_score = 0
        match_count=match_count+1
score_sum/match_count
```

### Global Results Table

```python
T20_noextra = T20_data.loc[(T20_data['extra_runs'] == 0) & (T20_data['is_wicket'] == 0)]
T20_extra = T20_data.loc[(T20_data['extra_runs'] > 0) & (T20_data['is_wicket'] == 0)]
T20_wicket = T20_data.loc[T20_data['is_wicket'] != 0]
```

```python
results_table = pd.DataFrame({'Outcome': pd.Series(dtype='str'),
                   'T20_prob': pd.Series(dtype='float'),
                   'ODI_prob': pd.Series(dtype='float')})
for i in range(7):
    #print("Probability of scoring " + str(i) + " run : " + str(T20_data['batsman_runs'].value_counts()[i]/T20_data.shape[0]*100))
    results_table = results_table.append({'Outcome': str(i), 'T20_prob': T20_noextra['total_runs'].value_counts()[i]/T20_data.shape[0]*100, 'ODI_prob' : ODI_data['Runs_Batsman'].value_counts()[i]/ODI_data.shape[0]*100},ignore_index=True)

results_table = results_table.append({'Outcome': 'extra runs', 'T20_prob': T20_extra.shape[0]/T20_data.shape[0]*100, 'ODI_prob': (ODI_data['Runs_Extras_Wides'].mean()+ODI_data['Runs_Extras_NoBall'].mean())*100},ignore_index=True)
results_table = results_table.append({'Outcome': 'wicket', 'T20_prob':T20_wicket.shape[0]/T20_data.shape[0]* 100, 'ODI_prob': ODI_data['Wicket_ball'].mean()},ignore_index=True)
results_table
```
