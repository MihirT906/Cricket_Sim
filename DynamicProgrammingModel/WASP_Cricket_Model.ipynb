```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_data = pd.read_csv('Cricket20Dataset.csv')
raw_data
```

```python
df = raw_data.copy()
columns_reqd = ['Balls', 'Wickets', 'Runs_Total', 'Wicket_ball', 'MatchNo', 'Score']
df = df[columns_reqd]
```

```python
raw_data.columns.values
```

```python
raw_data['Match_Type'].unique()
```

```python
df = df.dropna(axis=0)
df
```

```python
summary_table = df.groupby(['Balls', 'Wickets']).mean()#.agg(['mean', 'count'])
summary_table
```

```python
DP = np.zeros((8,4))
DP
```

```python
for i in range(6,0,-1):
    for j in range(2,-1,-1):
        if(i,j) not in summary_table.index:
            r_scored = 0.1
            p_w = 0.1
        else:
            r_scored = summary_table.loc[i,j].Runs_Total
            p_w = summary_table.loc[i,j].Wicket_ball
        
        DP[i,j] =  r_scored + p_w*DP[i+1,j+1] + (1-p_w)*DP[i+1,j]
        print("Balls:" + str(i) + " Wickets: "+ str(j) + " Value: " + str(DP[i,j]))
```

```python
print(DP)
```

```python
print("Predicted Score: " + str(DP[1,0]))
```
